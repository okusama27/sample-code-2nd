## 3.11 データの集計

```python
# リスト 3.11.1　DataFrame のグループ化

from urllib.parse import urljoin
import pandas as pd

base_url = "https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/"
anime_master_csv = urljoin(base_url, "anime_master.csv")
df = pd.read_csv(anime_master_csv)
grouped = df.groupby("type")
type(grouped)
```

```python
# リスト 3.11.2　get_group メソッドの実行

grouped.get_group("Movie").head()
```

```python
#  リスト 3.11.3　グループ化されたデータの個数

grouped.size()
```

```python
# リスト 3.11.4　グループ化されたデータの平均値

grouped.mean().round(1)
```

```python
# リスト 3.11.5　複数の要素で集約

df.groupby(["type", "episodes"]).mean().round(1).head(20)
```

```python
# リスト 3.11.6　numpy.median 関数で集計

import numpy as np

grouped.agg(np.median)
```

```python
# リスト 3.11.7　複数の関数で集計

grouped.agg([np.median, np.std])
```

```python
# リスト 3.11.8　列ごとに集計する関数を指定

grouped.agg({"episodes": np.median, "rating": np.mean})
```

```python
# リスト 3.11.9　pivot_table メソッドによる集計

df.pivot_table(index="type", aggfunc=np.mean)
```

```python
# リスト 3.11.10　pivot_table メソッドによる複数列の集計

df.pivot_table(index=["type", "episodes"], aggfunc=np.mean)
```

```python
# リスト 3.11.11　カンマ区切りの genre 列を 1 行ごとに分割①
unique_genres = np.sort(
    pd.Series(
        # 2次元から 1次元のデータに変換
        np.hstack(
            # genres列のカンマ区切りのデータを分割
            df["genre"]
            .str.split(",", expand=True)
            .values  # numpy.ndarrayのデータ型で取得
        )
    )
    .str.strip()  # 前後の空白文字を除去
    .dropna()  # 欠損値を除外
    .unique()  # 一意なデータに変換
)
```

```python
# リスト 3.11.12　genre 列ごとの DataFrame を結合

# 指定したジャンル名をDataFrameから抽出
def filter_df_by_genre(df, genre):
    genre_df = df.loc[df["genre"].str.contains(genre)].copy()
    genre_df["genre"] = genre
    return genre_df


# 上記の関数をすべてのジャンルに対して実行
genre_df_list = [filter_df_by_genre(df, genre) for genre in unique_genres]
# 上記dataを結合
tidy_df = pd.concat(genre_df_list)
# name列でソート
tidy_df.sort_values("name", inplace=True)
```

```python
# リスト 3.11.13　メンバ数上位 10 の genre 列のデータを抽出

# メンバ数が多いジャンルトップ10
top10 = tidy_df.groupby("genre")["members"].sum().nlargest(10).index
# top10からデータを抽出
tidy_top10_df = tidy_df.loc[tidy_df["genre"].isin(top10)]
```

```python
# リスト 3.11.14　複数のジャンルが登録された作品

# 「君の名は。」の抽出
tidy_top10_df.loc[tidy_top10_df["name"] == "Kimi no Na wa."]
```

```python
# リスト 3.11.15　クロス集計

tidy_top10_df.pivot_table(
    index="genre", columns="type", values=["members"], aggfunc=np.sum
).head()
```
