## 3.12: データの可視化

```python
# リスト 3.12.1　ノートブックにグラフを表示

%matplotlib inline

import pandas as pd

ax = pd.Series([1, 2, 3]).plot()
ax.set_title("Line Chart")
```

```python
# リスト 3.12.2　Series からグラフを出力

ser = pd.Series([1, 2, 3])
ax = ser.plot()
ax.set_title("Line Chart")
```

```python
# リスト 3.12.3　DataFrame からグラフを出力

df = pd.DataFrame({"a": [1, 2, 3], "b": [3, 2, 1]})
ax = df.plot()
ax.set_title("Line Chart")
```

```python
# リスト 3.12.4　折れ線グラフの描画

from urllib.parse import urljoin

base_url = "https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/"
anime_stock_returns_csv = urljoin(base_url, "anime_stock_returns.csv")
anime_stock_returns_df = pd.read_csv(anime_stock_returns_csv, index_col=0)

ax = anime_stock_returns_df.plot()
ax.set_title("stock returns")
```

```python
# リスト 3.12.5　Y 軸の範囲が異なるグラフを描画

anime_stock_price_csv = urljoin(base_url, "anime_stock_price.csv")
anime_stock_price_df = pd.read_csv(anime_stock_price_csv, index_col=0)
ax = anime_stock_price_df.plot(secondary_y=["IG Port"])
ax.set_title("secondary_y")
ax.set_ylabel("TOEI ANIMATION")
ax.right_ax.set_ylabel("IG Port")
```

```python
# リスト 3.12.6　複数のサブプロットに分割

ax1, ax2 = anime_stock_price_df.plot(subplots=True)
ax1.set_title("subplot1")
ax2.set_title("subplot2")
```

```python
# リスト 3.12.7　散布図の描画

anime_master_csv = urljoin(base_url, "anime_master.csv")
anime_master_df = pd.read_csv(anime_master_csv)
ax = anime_master_df.plot.scatter(x="members", y="rating")
ax.set_title("Scatter")
```

```python
# リスト 3.12.8　棒グラフの描画

anime_genre_top10_pivoted_csv = urljoin(
    base_url, "anime_genre_top10_pivoted.csv"
)
anime_genre_top10_pivoted_df = pd.read_csv(
    anime_genre_top10_pivoted_csv, index_col=0
)
ax = anime_genre_top10_pivoted_df.plot.bar()
```

```python
# リスト 3.12.9　対数軸の使用

ax = anime_genre_top10_pivoted_df.plot.bar(logy=True)
ax.legend(bbox_to_anchor=(1, 1))
```

```python
# リスト 3.12.10　積み上げ棒グラフの描画

ax = anime_genre_top10_pivoted_df.plot.bar(stacked=True)
ax.set_title("stacked")
```

```python
# リスト 3.12.11　pandas と seaborn を組み合わせたグラフ

import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
tips.groupby("day").mean().plot.bar(ax=ax1)
sns.violinplot(data=tips, y="tip", x="day", hue="smoker", split=True, ax=ax2)
ax1.legend(bbox_to_anchor=(0.6, 1.25))
ax2.legend(title="smoker", bbox_to_anchor=(0.6, 1.25))
```
