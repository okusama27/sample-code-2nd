## 5.2: フィギュアとサブプロットの書式設定

```python
# リスト 5.2.1　フィギュアタイトルの設定

import matplotlib.pyplot as plt

%matplotlib inline

fig, axes = plt.subplots(1, 2)
# フィギュアのタイトルの設定
fig.suptitle("フィギュアのタイトル", fontsize=20, fontweight="bold")
```

```python
# リスト 5.2.2　サブプロットタイトルの設定

fig, axes = plt.subplots(1, 2)

# フィギュアのタイトルの設定
fig.suptitle("フィギュアのタイトル")

# サブプロットタイトルの設定
axes[0].set_title("axes[0]のタイトル", {"fontsize": 16})
axes[1].set_title("axes[1]のタイトル", loc="right", pad=0)
```

```python
# リスト 5.2.3　フィギュアオブジェクトの引数設定

import matplotlib.pyplot as plt

%matplotlib inline

fig = plt.figure(
    figsize=(3, 3),  # サイズ
    dpi=100,  # 解像度
    facecolor="grey",  # 塗りつぶし色
    edgecolor="k",  # 枠線の色
    linewidth=10,  # 枠線の太さ
    frameon=True,  # 枠線を表示する
    tight_layout=False,
)
fig.add_subplot(1, 1, 1)
```

```python
# リスト 5.2.5　サブプロットの余白調整

fig, axes = plt.subplots(2, 2, linewidth=2, frameon=True)
fig.subplots_adjust(left=0.2, bottom=0.2, right=0.7, top=0.7, wspace=0.5, hspace=0.5)
fname = "subplots_adjust.png"
fig.savefig(fname, linewidth=2, edgecolor="grey")
```

```python
# リスト 5.2.6　出力したファイルを読み込んで確認

from IPython.display import Image

Image(fname)
```

```python
# リスト 5.2.7　サブプロットの書式設定

# 描画する値
y = [1, 10, 100, 1000, 10000]
x = range(len(y))
fig = plt.figure()

# サブプロットの書式設定
ax1 = fig.add_subplot(1, 2, 1, title="ax1", ylabel="linear values")
ax2 = fig.add_subplot(
    1,
    2,
    2,
    facecolor="lightblue",  # 塗りつぶし色
    title="ax2",  # サブプロットタイトル
    xticks=x,  # X軸目盛り
    xticklabels=list("abcde"),  # X軸目盛りラベル
    ylabel="log values",  # Y軸ラベル
    ylim=(1, 10000),  # Y軸範囲
    yscale="log",  # Y軸目盛りの形式
)
ax1.plot(x, y)
ax2.plot(x, y)
# サブプロット間のスペース調整
fig.subplots_adjust(wspace=0.4)
```
