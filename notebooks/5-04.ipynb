## 5.4: 凡例の設定

```python
# リスト 5.4.1　サブプロットに凡例を描画①

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
x = np.linspace(0, np.pi * 4, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig, ax = plt.subplots(1, 1)
ax.plot(x, y1, label="sin")
ax.plot(x, y2, label="cos")
# 凡例表示
ax.legend()
```

```python
# リスト 5.4.2　サブプロットに凡例を描画②

fig, ax = plt.subplots(1, 1)
ax.plot(x, y1)
ax.plot(x, y2)

# 凡例表示
ax.legend(["sin", "cos"])
```

```python
# リスト 5.4.3　サブプロットに凡例を描画③

fig, ax = plt.subplots(1, 1)
line1 = ax.plot(x, y1)[0]  # ①
line2 = ax.plot(x, y2)[0]  # ①

# 凡例表示
ax.legend((line1, line2), ("sin", "cos"))
```

```python
# リスト 5.4.4　サブプロットに凡例を描画

fig, axes = plt.subplots(1, 2, figsize=(9, 3))
axes[0].plot(x, y1, label="sin")
axes[1].plot(x, y2, label="cos")

# 凡例表示
axes[0].legend(loc="lower left")  # 左下に凡例を表示
axes[1].legend(loc=8)  # 水平方向中央下側に凡例を表示
```

```python
# リスト 5.4.5　bbox_to_anchor に値を指定して凡例を描画

fig, axes = plt.subplots(1, 2, figsize=(9, 3))
axes[0].plot(x, y1, color="blue", label="sin")
axes[1].plot(x, y2, color="red", label="cos")

# 凡例表示
axes[0].legend(loc="upper left", bbox_to_anchor=(2.2, 1.0))
axes[1].legend(loc="upper left", bbox_to_anchor=(1.0, 0.85))
```

```python
# リスト 5.4.6　borderaxespad の設定

fig, axes = plt.subplots(1, 2, figsize=(9, 3))
axes[0].plot(x, y1, color="blue", label="sin")
axes[1].plot(x, y2, color="red", label="cos")

# 凡例表示
axes[0].legend(loc="upper left", bbox_to_anchor=(2.22, 1.0), borderaxespad=0)
axes[1].legend(loc="upper left", bbox_to_anchor=(1.02, 0.85), borderaxespad=0)
```

```python
# リスト 5.4.7　フィギュアに凡例を設定

fig, axes = plt.subplots(1, 2, figsize=(9, 3))
axes[0].plot(x, y1, color="blue", label="sin")
axes[1].plot(x, y2, color="red", label="cos")

# 凡例表示
fig.legend()
```
