## 5.5: 線の書式設定

```python
# リスト 5.5.1　線の色と太さの設定

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
x = np.array([0, 1])
y = [0, 1]
fig, ax = plt.subplots(1, 1)

# 線の太さと不透明度を設定して描画
ax.plot(x, y, linewidth=1, color="b", label="linewidth=1(alpha=1.0)")
ax.plot(x + 0.1, y, linewidth=1, color="b", alpha=0.5, label="linewidth=1(alpha=0.5)")
ax.plot(x + 0.3, y, linewidth=10, color="b", label="linewidth=10(alpha=1.0)")
ax.plot(x + 0.4, y, linewidth=10, color="b", alpha=0.5, label="linewidth=10(alpha=0.5)")
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
```

```python
# リスト 5.5.2　線種を設定

s = np.array([0, 1])
t = np.array([14, 14])
fig, ax = plt.subplots()

# 線種を設定して描画
ax.plot(s, t, linewidth=5, color="blue", linestyle="solid", label="solid")
ax.plot(s, t - 1, linewidth=5, color="skyblue", linestyle="-", label="-")
ax.plot(s, t - 3, linewidth=5, color="blue", linestyle="dashed", label="dashed")
ax.plot(s, t - 4, linewidth=5, color="skyblue", linestyle="--", label="--")
ax.plot(s, t - 6, linewidth=5, color="blue", linestyle="dashdot", label="dashdot")
ax.plot(s, t - 7, linewidth=5, color="skyblue", linestyle="-.", label="-.")
ax.plot(s, t - 9, linewidth=5, color="blue", linestyle="dotted", label="dotted")
ax.plot(s, t - 10, linewidth=5, color="skyblue", linestyle=":", label=":")
ax.plot(
    s, t - 12, linewidth=5, color="orange", linestyle=(0, (5, 2)), label="(0, (5, 2))"
)
ax.plot(
    s, t - 13, linewidth=5, color="orange", linestyle=(4, (5, 2)), label="(4, (5, 2))"
)
ax.set_ylim(0, 15)
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
```

```python
#  リスト 5.5.3　線の先端形状の設定

x = [0, 1]
y = np.array([0.4, 0.4])
fig, ax = plt.subplots()

# 線の先端形状を設定して描画
ax.plot(x, y, linewidth=15, solid_capstyle="butt", label="butt")
ax.plot(x, y + 0.2, linewidth=15, solid_capstyle="round", label=" round")
ax.plot(x, y + 0.4, linewidth=15, solid_capstyle="projecting", label=" projecting")
ax.set_ylim([0, 1])
ax.legend(loc="lower right")
```

```python
# リスト 5.5.4　破線・点線の先端形状の設定

fig, ax = plt.subplots()

# 破線・点線の先端形状を設定して描画
ax.plot(x, y, linewidth=15, linestyle="-.", dash_capstyle="butt", label="butt")
ax.plot(x, y + 0.2, linewidth=15, linestyle="-.", dash_capstyle="round", label=" round")
ax.plot(
    x,
    y + 0.4,
    linewidth=15,
    linestyle="-.",
    dash_capstyle="projecting",
    label=" projecting",
)
ax.set_ylim([0, 1])
ax.legend()
```

```python
# リスト 5.5.5　線の結合点の形状を設定

p = np.linspace(0, 4, 4)
q = [0, 1, 0, 1]
fig, ax = plt.subplots(figsize=(8, 3))

# 線の結合点の形状を設定して描画
ax.plot(p, q, linewidth=10, solid_joinstyle="miter", label="miter")
ax.plot(p + 0.5, q, linewidth=10, solid_joinstyle="round", label="round")
ax.plot(p + 1.0, q, linewidth=10, solid_joinstyle="bevel", label="bevel")
ax.set_ylim(-0.1, 1.1)
ax.legend(loc="lower right")
```

```python
# リスト 5.5.6　破線の結合形状の設定

fig, ax = plt.subplots(figsize=(8, 3))

# 破線の結合形状を変えて描画
ax.plot(p, q, linewidth=10, linestyle="--", dash_joinstyle="miter", label="miter")
ax.plot(
    p + 0.5, q, linewidth=10, linestyle="--", dash_joinstyle="round", label="round",
)
ax.plot(
    p + 1.0, q, linewidth=10, linestyle="--", dash_joinstyle="bevel", label="bevel",
)
ax.set_ylim(-0.1, 1.1)
ax.legend(loc="lower right")
```
