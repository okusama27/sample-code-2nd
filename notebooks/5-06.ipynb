## 5.6: 色の設定

```python
# リスト 5.6.1　色の設定

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
x = np.arange(0, 6, 0.1)
y = np.sin(x)
z = np.cos(x)
fig, ax = plt.subplots(1, 1)

# 描画
ax.scatter(x, y, label="sin", color="lightblue", edgecolor="blue")
ax.scatter(x, z, label="cos", color="lightblue", edgecolor="blue", alpha=0.3)
ax.legend(loc="lower left")
```

```python
# リスト 5.6.2　色指定の例

fig, ax = plt.subplots(1, 1)
v = np.array([0, 1])

# 各種方法による色指定で描画
ax.plot(v, v, color="k", label="color string", linewidth=8)
ax.plot(v + 1, v, color="0.5", label="Gray scale", linewidth=8)
ax.plot(v + 2, v, color=(0, 0, 0.7, 1), label="RGBA tuple", linewidth=8)
ax.plot(v + 3, v, color="#66b3e9", label="RGBA 16", linewidth=8)
ax.plot(v + 4, v, color="DarkGreen", label="HTML color", linewidth=8)
ax.plot(v + 5, v, color="LimeGreen", label="X11/CSS", linewidth=8)
ax.plot(v + 6, v, color="C3", label="Regular expression", linewidth=8)
ax.plot(v + 7, v, color="tab:pink", label="Tableau", linewidth=8)
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
```

```python
# リスト 5.6.3　カラーマップを取得して色を設定

import matplotlib.cm as cm

fig, ax = plt.subplots(1, 1)
cname = "viridis"
n = 10
cmap = cm.get_cmap(cname, n)
for i in range(n):
    ax.plot(x, y + n - i, color=cmap(i), label=f"{cname}({i})")

ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
```

```python
# リスト 5.6.4　カラーマップを取得して色を設定

fig, ax = plt.subplots(1, 1)
cmap = cm.viridis
for i in np.linspace(0, 1, 10):
    ax.plot(x, y + 2 - i, color=cmap(i), label=f"viridis({i:.1f})")

ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
```

```python
# リスト 5.6.5　段階的に色を変化させた散布図

fig, ax = plt.subplots(1, 1)
ax.scatter(x, y, c=y, cmap=cm.Blues)
```

```python
# リスト 5.6.6　個別に色を指定した散布図

fig, ax = plt.subplots(1, 1)
x = [0, 1]
y = np.array([0.4, 0.4])
cmap = cm.get_cmap("viridis")
ax.plot(x, y, color=cmap(0.0), label="cmap(0.0)")
ax.plot(x, y - 0.2, color=cmap(0.6), label="cmap(0.6)")
ax.set_ylim(0, 0.6)
ax.legend()
```
