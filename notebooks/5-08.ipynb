## 5.8: 描画スタイルの適用

```python
# リスト 5.8.1　スタイル一覧の出力

import matplotlib.pyplot as plt

plt.style.available
```

```python
# リスト 5.8.2　ggplot スタイルで描画

import matplotlib
import numpy as np

np.random.seed(12345)
x1 = np.random.normal(100, 10, 1000)
x2 = np.random.normal(150, 10, 1000)
y1 = np.random.normal(100, 50, 1000)
y2 = np.random.normal(50, 50, 1000)
matplotlib.style.use("ggplot")
fig, ax = plt.subplots(1, 1)
ax.scatter(x1, y1)
ax.scatter(x2, y2)
```

```python
# リスト 5.8.3　grayscale スタイルで描画

matplotlib.style.use("grayscale")
fig, ax = plt.subplots(1, 1)
ax.scatter(x1, y1)
ax.scatter(x2, y2)
```
