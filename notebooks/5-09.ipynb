## 5.9: グラフをファイルに出力

```python
# リスト 5.9.1　フィギュアオブジェクトの書式を設定

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fname = "save_fig_settings.png"
fig.savefig(fname, dpi=300, facecolor="lightgrey", edgecolor="k", format="png")
```

```python
# リスト 5.9.2　リスト 5.9.1 で出力したファイルを読み込み

from IPython.display import Image

Image(fname)
```

```python
# リスト 5.9.3　bbox_inches に "tight" を設定して画像を出力

import numpy as np

x = np.arange(0, 5, 0.2)
y = np.sin(x)
z = np.cos(x)
fig, ax = plt.subplots(1, 1)
ax.plot(x, y, label="sin")
ax.plot(x, z, label="cos")
ax.set_xlabel("x value")
ax.set_ylabel("y value")
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
fname = "save_fig_adjust.png"
fig.savefig(
    fname, dpi=300, bbox_inches="tight",
)
```
